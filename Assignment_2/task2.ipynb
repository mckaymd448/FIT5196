```python
# Code to audit data
#import the second dataset into python as a dataframe.
csv_dataset2 = pd.read_csv("dataset2.csv")

#Inspect the shape of dataFrame, and view the top 10 rows.
print(csv_dataset2.shape)
csv_dataset2.head(10)
```

```python
#Start renaming columns so they match the first dataset.

csv_dataset2 = csv_dataset2.rename(columns = {'Job Title':'Title'})
csv_dataset2 = csv_dataset2.rename(columns = {'Organisation':'Company'})

# convert salary / month to salary / year and del Salary per month column.
csv_dataset2['Salary'] = csv_dataset2['Salary per month'].apply(lambda x: x * 12)
del csv_dataset2['Salary per month']

csv_dataset2 = csv_dataset2.rename(columns = {'Contract Type':'ContractTime'})
csv_dataset2 = csv_dataset2.rename(columns = {'Fraction':'ContractType'})

csv_dataset2.head(10)
```

```python
# Convert the syntax of the Opening and Closing columns to match the previous file.

def reformatDate(inp_str):
    if pd.isnull(inp_str):
        return inp_str
    else:
        inp_list = inp_str.split(' ')
        date_str = inp_list[0]
        time_str = inp_list[1]
        out_str = date_str + 'T' + time_str
        out_str = re.sub(r"[-:]+", "", out_str)
        return out_str

csv_dataset2['OpenDate'] = csv_dataset2['Opening'].apply(lambda x: reformatDate(x))
csv_dataset2['CloseDate'] = csv_dataset2['Closing'].apply(lambda x: reformatDate(x))

csv_dataset2 = csv_dataset2.drop('Opening', axis=1)
csv_dataset2 = csv_dataset2.drop('Closing', axis=1)

csv_dataset2.head(10)
```

```python
# Insert empty columns for Id and Source.
csv_dataset2['Id'] = ''
csv_dataset2['Source'] = 'csv_dataset2'

csv_dataset2.head(10)
```

```python
# reorder columns to match the previous input file.
csv_dataset2 = csv_dataset2[['Id','Title','Location','Company','ContractType','ContractTime','Category','Salary','OpenDate','CloseDate','Source']]
csv_dataset2.head(10)
```

```python
# check datatypes in the new csv_dataset2 dataFrame.
print(csv_dataset2.info())
```

```python
# change 'Id' to integer and 'Salary' to a float datatype like the previous format.
csv_dataset2['Id'] = pd.to_numeric(csv_dataset2['Id'])
csv_dataset2['Salary'] = pd.to_numeric(csv_dataset2['Salary'], downcast = "float")

csv_dataset2.info()
```

```python
# Open the datafile created by task1.
csv_dataset = pd.read_csv("dataset1_solution.csv")
print(csv_dataset.shape)
```

```python
# Determine what the maximum ID number was for ID in the first dataset.
csv_dataset['Id'].max()
```

```python
# The ID for the second dataset should be an increment from 80000000 onwards to be safe.
csv_dataset2['Id'] = csv_dataset2.index + 80000000
csv_dataset2.head()
```

```python
# Combine the two datasets together.
csv_dataset = csv_dataset.append(csv_dataset2)
print(csv_dataset.shape)
```

```python
# Check what the category names are, and how they compare between datasets.
csv_dataset['Category'].unique()
```

```python
# Update categories so they are there are not duplicate categories.

def updateCategories(inp_str):
    if pd.isnull(inp_str):
        return inp_str
    else:
        inp_str = inp_str.replace('PR, Advertising and Marketing', 'PR, Advertising & Marketing Jobs')
        inp_str = inp_str.replace('Finance and Accounting', 'Accounting & Finance Jobs')
        inp_str = inp_str.replace('Information Technology', 'IT Jobs')
        return inp_str
    
csv_dataset['Category'] = csv_dataset['Category'].apply(lambda x: updateCategories(x))

csv_dataset['Category'].unique()
```

```python
# Check what the Contract Types are, and how they compare between datasets.
csv_dataset['ContractType'].unique()
```

```python
# Normalise the contract type category so they all match.  Part Time appears twice.  Combine into one category
# called part_time.

def updateContractType(inp_str):
    if pd.isnull(inp_str):
        return inp_str
    else:
        inp_str = inp_str.replace('Part Time', 'part_time')
        return inp_str
    
csv_dataset['ContractType'] = csv_dataset['ContractType'].apply(lambda x: updateContractType(x))

csv_dataset['ContractType'].unique()
```

```python
# Check what the Contract Times are, and how they compare between datasets.
csv_dataset['ContractTime'].unique()
```

```python
# Update categories so they are there are not duplicate categories.

def updateContractTime(inp_str):
    if pd.isnull(inp_str):
        return inp_str
    else:
        inp_str = inp_str.replace('Contract', 'contract')
        inp_str = inp_str.replace('Permanent', 'permanent')
        return inp_str
    
csv_dataset['ContractTime'] = csv_dataset['ContractTime'].apply(lambda x: updateContractTime(x))

csv_dataset['ContractTime'].unique()
```

```python
# Check for duplicates in the final dataset using the Title, Location and Company fields.

dup_check = csv_dataset.loc[csv_dataset.duplicated(subset=['Title','Location','Company']), :]

print('There are',len(dup_check),' duplicate items.')

dup_check_list = list(dup_check['Title'])

dup_csv_data = csv_dataset[csv_dataset.Title.isin(dup_check_list)]

# View the dataframe, sorted by Title.
dup_csv_data.sort_values(by=['Title'])
```

```python
# Remove the duplicated values.
csv_dataset = csv_dataset.drop_duplicates(subset=['Title','Location','Company'], keep='first')
```
